 ## UNIVARIATE ANALYSIS 
 <font face="Times New Roman" size="4" > <em> Univariate Exploratory Analysis Involves Examining and Understanding One Variable at a Time. Instead of Looking at Relationships Between Variables, Univariate Analysis Focuses on Exploring the Characteristics and Patterns of Individual Variables in Isolation. </em></font>

```python
plt.figure(figsize=(6,6))
sns.countplot(x = 'Customer Segment' ,data = df)
plt.gcf().set_facecolor('#f0f0f0')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x="Product Category", data=df)
plt.gcf().set_facecolor('#f0f0f0')
plt.show()
```

```python
x= df['Product Sub-Category'].value_counts().index
y= df['Product Sub-Category'].value_counts().values
```

```python
plt.figure(figsize=(50,25))
plt.pie(y, labels=x,startangle=60,autopct='%1.1f%%')
plt.rcParams['font.size'] = 40
plt.title(" Product Sub Category")
plt.legend(x)
plt.legend(fontsize = 20, loc= "lower left")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x="Ship Mode", data = df)
plt.gcf().set_facecolor('#f0f0f0')
#plt.rcParams['font.size'] = 15
plt.show()
```

```python
plt.figure(figsize=(25,10))
sns.countplot(x='Product Container',data=df)
plt.rcParams['font.size'] = 15
plt.gcf().set_facecolor('#f0f0f0')
plt.show()
```

```python
plt.figure(figsize=(35,40))
top_10_values = df['State or Province'].value_counts().head(10)
x= top_10_values.index
y= top_10_values.values
plt.pie(y, labels=x, startangle=20,autopct='%1.1f%%' )
plt.rcParams['font.size'] = 30
plt.title("Top 10 State Rankings With Respect To Purchasing Power")
plt.legend(x)
plt.legend(fontsize='30', loc='upper left')
plt.show()
```

```python
last_10 = df['State or Province'].value_counts().tail(10)
x= last_10.index
y= last_10.values
plt.figure(figsize=(35,40))
plt.pie(y, labels=x,startangle=20,autopct='%1.1f%%' )
plt.rcParams['font.size'] = 30
plt.legend(x)
plt.legend(fontsize=30, loc = 'upper left' )
plt.title("Bottom  10 State Rankings With Respect To Purchasing Power")
plt.show()
```

```python
df['year'] = df["Order Date"].dt.year
plt.figure(figsize=(12,8))
sns.countplot(x='year', data=df)
plt.gcf().set_facecolor('#f0f0f0')
plt.title("Sales Per Year")
plt.show()
```

```python
plt.figure(figsize=(12,8))
df['year'] = df["Order Date"].dt.year
ax = sns.countplot(x='year', data=df, hue='Product Category', palette = "Set1")
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom',  # Adjust va to 'bottom' for count below bars
                fontsize=12, color='black', 
                xytext=(0, 5), 
                textcoords='offset points')
plt.gcf().set_facecolor('#f0f0f0')
plt.legend(fontsize=15, loc = 'upper left' )
plt.title('Count of Orders by Year and Product Category')

plt.show()
```
